Soal 3: Tranformasikan data tersebut,dengan harapan saat divisualisasi  NA_Sales,EU_Sales,JP_Sales dan Other_Sales ada di sumbu-x

```python
import pandas as pd

# Baca data
df = pd.read_csv("vgsales.csv")  # ganti path kalau perlu

# Kolom yang akan dihitung rata-ratanya
sales_cols = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']

# Group by Genre
avg_sales = df.groupby('Genre')[sales_cols].mean()

# Transpose hasil supaya Region jadi index
transposed = avg_sales.T.reset_index()
transposed = transposed.rename(columns={'index': 'Region'})

# Melt biar jadi long format (Region di kolom, Genre & Sales sebagai pasangan)
melted = transposed.melt(id_vars='Region', var_name='Genre', value_name='Average_Sales')

# Tampilkan hasil akhir
print(melted.head())
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>
1. Visualisasikan gambar yang tadi kamu manipulasi, untuk mengjasilkan visualisasi seperti di bawah ini
2. Tuliskan kesimpulan yang didapat dari visualisasi tersebut

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Baca data CSV
df = pd.read_csv("vgsales.csv")

# Group dan hitung rata-rata penjualan regional per genre
sales_cols = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
avg_sales = df.groupby('Genre')[sales_cols].mean()

# Transpose biar region jadi sumbu-X
transposed = avg_sales.T.reset_index()
transposed = transposed.rename(columns={'index': 'Region'})

# Ubah jadi long format
melted = transposed.melt(id_vars='Region', var_name='Genre', value_name='Average_Sales')

# Buat visualisasi
plt.figure(figsize=(14, 7))
sns.barplot(data=melted, x='Region', y='Average_Sales', hue='Genre')
plt.title('Perbandingan Rata-Rata Penjualan Game per Genre di Tiap Region 🎮🌍')
plt.ylabel('Rata-rata Penjualan (Juta Unit)')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

<h2>Soal 5: Perbandingan Kuantitatif Barplot - Stack Barplot</h2>

1. Visualisasikan gambar yang tadi kamu manipulasi, untuk mengjasilkan visualisasi seperti di bawah ini
2. Tuliskan kesimpulan yang didapat dari visualisasi tersebut

```python
import pandas as pd
import matplotlib.pyplot as plt

# Baca data CSV
df = pd.read_csv("vgsales.csv")

# Group dan ambil rata-rata per genre per region
sales_cols = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
avg_sales = df.groupby('Genre')[sales_cols].mean()

# Transpose biar region jadi index (perlu buat stacked bar)
stacked_data = avg_sales.T  # rows = Region, columns = Genre

# Visualisasi stacked barplot
plt.figure(figsize=(14, 7))
stacked_data.plot(kind='bar', stacked=True, colormap='tab20', figsize=(14, 7))

plt.title('Stacked Barplot: Rata-rata Penjualan Game per Genre di Tiap Region 🎮📦')
plt.ylabel('Rata-rata Penjualan (Juta Unit)')
plt.xlabel('Region')
plt.xticks(rotation=0)
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
