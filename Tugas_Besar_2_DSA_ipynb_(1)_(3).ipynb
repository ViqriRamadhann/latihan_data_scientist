<h2>Soal 5:Apa itu Deskriptif Statistik?</h2>



**jawab**: Deskriptif Statistik adalah cabang statistik yang bertujuan untuk menggambarkan dan merangkum karakteristik utama dari suatu kumpulan data. Fokus utama dari deskriptif statistik adalah untuk menyajikan data dalam bentuk yang mudah dipahami tanpa melakukan analisis yang lebih kompleks. Teknik-teknik yang digunakan meliputi pengukuran tendensi sentral seperti rata-rata (mean), median, dan modus yang memberikan gambaran tentang nilai pusat dalam data. Selain itu, ada pengukuran dispersi atau penyebaran data seperti rentang (selisih antara nilai maksimum dan minimum), variansi, dan standar deviasi untuk melihat sejauh mana data tersebar dari rata-rata. Deskriptif statistik juga sering menggunakan visualisasi data, seperti diagram batang, histogram, dan boxplot, untuk memudahkan pemahaman dan interpretasi data. Secara keseluruhan, deskriptif statistik membantu kita untuk mendapatkan gambaran umum dan pola dalam data secara sederhana dan jelas.
